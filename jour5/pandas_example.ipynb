```python
import pandas as pd

df = pd.read_csv("train.csv")
df
# df['Name']
# df.columns
# df.describe()

df['Ticket']
```

```python
import re
import numpy as np

def extract_number(ticket):
    result = re.findall('\d+$', ticket)
    if len(result):
        return result[0]
    return np.nan
df['TicketClean'] = df['Ticket'].map(extract_number)

#Filtre
df[df['Cabin'].isna()]
```
